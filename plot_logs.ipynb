```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from numpy import genfromtxt

drone_log_path = './build/drone_log.txt'
scene_log_path = './build/scene_log.txt'
scene_laser_log_path = './build/scene_laser_log.txt'
drone_log = genfromtxt(drone_log_path, delimiter=',')
scene_log = genfromtxt(scene_log_path, delimiter=',')
scene_laser_log = genfromtxt(scene_log_path, delimiter=',')
```

```python
# Drone log format height, perceived height, laser height, velocity z
# print(drone_log[:, 0])
fig = plt.figure('drone_log')
plt.plot(drone_log[:,0])
plt.plot(drone_log[:,1])
plt.plot(drone_log[:,2])
plt.plot(scene_log[:,2])
# plt.plot(np.arange(0, len(scene_laser_log))/len(scene_laser_log)*len(drone_log), scene_laser_log)

plt.legend(['height', 'perceivedheight', 'laser height', 'Truth', 'laser truth'])
plt.title('Drone height logs')
plt.show()

fig2 = plt.figure('vel_log')
plt.plot(drone_log[:,3])
plt.legend(['velocity'])
plt.title('Drone velocity logs')
plt.show()
```

```python
fig3 = plt.figure('3D')

plt.rcParams["figure.autolayout"] = True

line = np.linspace(0, 10, 100)
x, y = np.meshgrid(line, line)
z = x*0

fig = plt.figure()

ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z)
ax.plot(scene_log[:,0], scene_log[:,1], scene_log[:,2])

# ax.voxels(bx.plt_points())
plt.show()
```
